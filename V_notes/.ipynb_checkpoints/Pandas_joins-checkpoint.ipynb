```python
df1 = pd.DataFrame({"Name":["Akshay",'Swapnil','Mangesh','vijay','Sachin'],
                   "Age":[20,24,35,25,40],
                   "Location":['Pune','Mumbai','Banglore','Delhi','Kolkata']})

df2 = pd.DataFrame({"Name":["Ajay",'Akshay','Pravin',"Suraj",'Sachin'],
                   "Age":[40,20,39,18,40],
                   "Location":['Chennai','Pune','Banglore','Delhi',"Kolkata"],
                   "Salary":[10000,40000,50000,30000,90000]})
df1
```

```python
df2
```

```python
df1.join(df2,how = "outer",lsuffix = "_df1",rsuffix="_df2")

#lsuffix : str, default ''
#Suffix to use from left frame's overlapping columns.
#rsuffix : str, default ''
#Suffix to use from right frame's overlapping columns.
```

```python
df1.join(df2,how = "inner",lsuffix = "_df1",rsuffix="_df2")
```
